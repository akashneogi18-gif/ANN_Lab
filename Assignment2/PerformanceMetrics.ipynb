```python
import numpy as np
import matplotlib.pyplot as plt, seaborn as sns
class PerformanceMetrics:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred
        self.classes = np.unique(y_true)
        self.class_to_index = {c: i for i, c in enumerate(self.classes)}

    def confusion_matrix(self):
        n = len(self.classes)
        cm = np.zeros((n, n), dtype=int)
        for t, p in zip(self.y_true, self.y_pred):
            cm[t][p] += 1
        return cm

    def accuracy(self):
        correct = np.sum(self.y_true == self.y_pred)
        total = len(self.y_true)
        return correct/total

    def precision(self):
        cm = self.confusion_matrix()
        precisions = []
        for i in range(len(self.classes)):
            tp = cm[i][i]
            fp = cm[:, i].sum() - tp
            if tp+fp == 0:
                precisions.append(0)
            else:
                precisions.append(tp/(tp+fp))
        return np.mean(precisions)

    def recall(self):
        cm = self.confusion_matrix()
        recalls = []
        for i in range(len(self.classes)):
            tp = cm[i][i]
            fn = cm[i, :].sum() - tp
            if tp + fn == 0:
                recalls.append(0)
            else:
                recalls.append(tp/(tp+fn))
        return np.mean(recalls)

    def f1_score(self):
        p = self.precision()
        r = self.recall()
        if p+r == 0:
            return 0
        return 2 * (p * r) / (p + r)

    def report(self):
        return {
            "Accuracy" : f"{self.accuracy():.3f}",
            "Precision" : f"{self.precision():.3f}",
            "Recall" : f"{self.recall():.3f}",
            "F1 Score" : f"{self.f1_score():.3f}"
        }
        
    def plot_confusion_matrix(self):
        cm = self.confusion_matrix()
        plt.figure(figsize=(6,5))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                    xticklabels=self.classes, yticklabels=self.classes)
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.title("Confusion Matrix")
        plt.show()
```
